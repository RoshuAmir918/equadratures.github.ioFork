## Polynomial over cubes

We demonstrate the use of ``equadratures`` for fitting polynomials across lines, squares, and more generally hypercubes. 

```python
from equadratures import *
import matplotlib.pyplot as plt
```

For illustrative purposes consider, the function below

```python
xi = np.linspace(0, 2*np.pi, 1500)
def fun(x):
    return np.sin(50*x) * np.exp(-2*x)

# plot!
fig = plt.figure(figsize=(10,4))
plt.plot(xi, fun(xi))
plt.show()
```

Using a small number of samples, our goal is to arrive at the same resolution as the function above. To begin, we define the support, i.e.,

```python
x = Parameter(lower=0, upper=2*np.pi, order=200)
b = Basis('univariate')
poly = Poly(x, b, method='numerical-integration')
poly.set_model(fun)
yp = poly.get_polyfit(xi)
```

```python
fig = plt.figure(figsize=(10,4))
plt.plot(xi, fun(xi), alpha=0.5)
plt.plot(poly._quadrature_points, poly._model_evaluations, '.', ms=10)
plt.plot(xi, yp, '-')
plt.show()
```

```python
fig = plt.figure(figsize=(10,4))
plt.plot(xi, fun(xi) - yp.flatten(), 'k=')
plt.title('Error')
plt.show()
```
